```python
# read image by PIL
# with 可以在段落結束後，關閉image 保持資料正常使用
with Image.open("demo.jpg") as img:
    plt.imshow(img)
    plt.show()
```

## 將圖轉換成array 在轉換回來

```python
df = pd.DataFrame({})

with Image.open("demo.jpg") as img:
    
    print(img.size)
    print("--------------")
    
    # method 1
    imgData = img.getdata() # getdata flattens the pixels into a 1D list
    print("--------------")
    
    im2arr0 = np.array(imgData)
    im2arr0 = im2arr0.reshape((img.size[1], img.size[0], 3))
    print(im2arr0.dtype)
    print(im2arr0.shape)
    
    newimg0 = Image.fromarray(im2arr0, "RGB")
    plt.imshow(newimg0)
    plt.show()
    
    # method 2
    print("--------------")
    im2arr1 = np.array(img) # im2arr.shape: height x width x channel
    print(im2arr1.dtype)
    print(im2arr1.shape)
    
    newimg = Image.fromarray(im2arr1, "RGB")
    plt.imshow(newimg)
    plt.show()
    
    # check
    print(type(im2arr0))
    print(type(im2arr1))
    print(np.array_equal(im2arr0, im2arr1))
    
    # save to pd
    df["NG"] = [im2arr0]
    df["OK"] = [im2arr1]
```

### [ISSUE] 兩種方式產生的 array 數值一樣可是還原後的圖不一樣

data type不同，雖然都是int 但一個是int64 一個是int8，fromarray要input int8，其內部的解碼才會正確拼出圖

## using ndimage from scipy

mode{‘reflect’, ‘constant’, ‘nearest’, ‘mirror’, ‘wrap’}, optional 

The mode parameter determines how the input array is extended beyond its boundaries. Default is ‘reflect’. Behavior for each valid value is as follows:

‘reflect’ (d c b a | a b c d | d c b a)

The input is extended by reflecting about the edge of the last pixel. This mode is also sometimes referred to as half-sample symmetric.

-

‘constant’ (k k k k | a b c d | k k k k)

The input is extended by filling all values beyond the edge with the same constant value, defined by the cval parameter.

-

‘nearest’ (a a a a | a b c d | d d d d)

The input is extended by replicating the last pixel.

-

‘mirror’ (d c b | a b c d | c b a)

The input is extended by reflecting about the center of the last pixel. This mode is also sometimes referred to as whole-sample symmetric.

-

‘wrap’ (a b c d | a b c d | a b c d)

The input is extended by wrapping around to the opposite edge.

```python
# use scipy
from scipy import ndimage

img = df["OK"][0]
imgarr = np.array(img)

k = 10
kMatrix = (1 / (k**2)) * np.ones((k, k))

for d in range(len(imgarr[0,0])):

    res = imgarr[:, :, d].copy()
    res = ndimage.convolve(res.astype("float64"), kMatrix, mode='constant', cval=0.0)
    
    print(res)
    # rescale
    res[res > 255] = 255
    res[res < 0] = 0
    imgarr[:, :, d] = res.astype("uint8")

print("*********************")
print(imgarr.dtype)

conv_img = Image.fromarray(imgarr)

plt.figure(figsize=(16, 12))
plt.imshow(conv_img)
plt.show()
```
